```python
df = pd.read_csv('labeled_reviews.csv')
df
```

```python
idx = int(input("Starting idx:"))
while True:
    print("\n" + str(idx) + " " + df.iloc[idx]['Review'])
    ch = input(": ")
    if ch == 'q': 
        break
    elif ch == '1':
        df.at[idx , 'isSuggestion'] = True
    else:
        df.at[idx, 'isSuggestion'] = False
    idx = idx + 1
```
